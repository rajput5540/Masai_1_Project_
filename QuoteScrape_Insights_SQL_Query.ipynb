```python
z = 'Select * from Book_Description limit 5'
pd.read_sql(z,engine)
```

```python
x = ''' select Tags, count(*) as tag_count
        from Book_Description
        Group by Tags
        Order by count(Tags) desc
        limit 5'''
pd.read_sql(x,engine)
```

```python
y = ''' select Author, count(*) as quotes_count
        from Book_Description
        Group by Author
        Having count(*) > 5'''
pd.read_sql(y,engine)
```

```python
x = ''' select *, length(Quote) as quote_len
        from Book_Description
        Order by length(Quote) desc
        limit 1'''
pd.read_sql(x,engine)
```
